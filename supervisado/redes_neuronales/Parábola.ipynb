```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
rng = np.random.RandomState(54321)

x_data = rng.uniform(-10,10,500) 
y_data = x_data*x_data + 3*rng.randn(500)

x_data=x_data[:, np.newaxis]
y_data=y_data[:, np.newaxis]
```

```python
plt.scatter(x_data, y_data)
x_new = np.linspace(-10, 10)
plt.plot(x_new, x_new*x_new, c='green')
plt.show()
```

```python
scaler = MinMaxScaler() 

x_scaled = scaler.fit_transform(x_data)
```

## Sigmoide

```python
tf.random.set_seed(1234)
```

```python
def create_model():
    return tf.keras.models.Sequential([
        tf.keras.layers.Dense(50, input_dim=1, activation='sigmoid'),
        tf.keras.layers.Dense(1, activation='linear')
    ])
```

```python
model = create_model()
model.compile(optimizer='sgd',
              loss='mse',
              metrics=['mse'])

history = model.fit(x=x_scaled, 
          y=y_data, 
          epochs=250,
          verbose=0)
```

```python
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper left')
plt.show()
```

```python
x_new = np.linspace(-10, 10)
x_new = x_new[:, np.newaxis]

x_new_scaled = scaler.transform(x_new)
y_new = model.predict(x_new_scaled)
```

```python
plt.scatter(x_data, y_data)
plt.plot(x_new, y_new, c='red')
```

```python
x_new = np.linspace(-15, 15)
x_new = x_new[:, np.newaxis]

x_new_scaled = scaler.transform(x_new)
y_new = model.predict(x_new_scaled)

plt.scatter(x_data, y_data)
plt.plot(x_new, y_new, c='red')
```
